```python
from covid import Covid
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
covid = Covid(source='worldometers')
covid
```

```python
data = covid.get_data()
data
```

```python
covid_df = pd.json_normalize(data)
covid_df[20:27]
```

```python
plt.style.available
```

```python
plt.style.use('classic')

fig, (ax0, ax1, ax2, ax3) = plt.subplots(
    figsize=(30, 50),
    nrows=4,
    ncols=1,
)

fig.suptitle("Covid-19 Live Analysis", fontsize=50)

ax0.bar(covid_df["country"][8:30], covid_df["confirmed"][8:30])
ax0.set_title("Total Confirmed Cases (By Countries)", fontsize=30)
ax0.set_xlabel("Countries", fontsize=20)
ax0.set_ylabel("Confirmed Cases", fontsize=20)

ax1.bar(covid_df["country"][8:30], covid_df["deaths"][8:30])
ax1.set_title("Total Confirmed Deaths (By Countries)", fontsize=30)
ax1.set_xlabel("Countries", fontsize=20)
ax1.set_ylabel("Confirmed Deaths", fontsize=20)

ax2.bar(covid_df["country"][8:30], covid_df["recovered"][8:30])
ax2.set_title("Total Confirmed Recovered Cases (By Countries)", fontsize=30)
ax2.set_xlabel("Countries", fontsize=20)
ax2.set_ylabel("Recovered Cases", fontsize=20)

ax3.bar(covid_df["country"][8:30], covid_df["active"][8:30])
ax3.set_title("Total Active Cases (By Countries)", fontsize=30)
ax3.set_xlabel("Countries", fontsize=20)
ax3.set_ylabel("Active Cases", fontsize=20)

fig.savefig("covid-19_live_analysis.png")
```
